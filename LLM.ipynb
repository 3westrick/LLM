## Day 4 | POS Tagging & NER

```python
# Load the pre-trained English model
# The 'nlp' object is now your gateway to all of spaCy's power.
nlp = spacy.load("en_core_web_sm")

# Here's our sample text with a few entities
my_text = "Sundar Pichai, the CEO of Google, announced the new Gemini model from London last week."

# Process the text with the spaCy pipeline
doc = nlp(my_text)

# 3. Perform Part-of-Speech (POS) Tagging
print("--- PART-OF-SPEECH TAGS ---")
# The 'doc' object is an iterable of 'Token' objects.
# Each token has rich linguistic annotations.
for token in doc:
    # token.text is the word itself
    # token.pos_ is the simple part-of-speech tag
    # token.tag_ is the more detailed tag
    print(f"{token.text:<12} | {token.pos_:<10} | {spacy.explain(token.pos_)}")

# 4. Perform Named Entity Recognition (NER)
print("\n--- NAMED ENTITIES ---")
# The 'doc' object also has an 'ents' attribute with all found entities.
for ent in doc.ents:
    # ent.text is the entity's text (e.g., "Sundar Pichai")
    # ent.label_ is the entity's category (e.g., "PERSON")
    print(f"{ent.text:<20} | {ent.label_:<10} | {spacy.explain(ent.label_)}")
```

## Day 6 | First Deep Learning NLP Model in PyTorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import gensim.downloader as api

# --- Part 1: Load Word Vectors and Prepare Data ---

# Load the same Word2Vec model from yesterday
# This will be fast if it's already downloaded
print("Loading Word2Vec model...")
wv = api.load('word2vec-google-news-300')
print("Model loaded.")

# Our simple dataset [("sentence", label), ...]
# 0 for Negative, 1 for Positive
train_data = [
    ("This movie was terrible and a waste of time.", 0),
    ("I absolutely loved this, it was fantastic!", 1),
    ("The acting was poor and the plot was boring.", 0),
    ("This is the best film I have seen all year.", 1),
    ("A truly moving and brilliant masterpiece.", 1),
    ("What a horrible, dreadful experience.", 0)
]

# --- Part 2: Function to Vectorize a Sentence ---

def sentence_to_vector(sentence, word_vectors):
    """Converts a sentence to a single vector by averaging its word embeddings."""
    words = sentence.lower().split()
    word_vecs = [word_vectors[word] for word in words if word in word_vectors]

    if not word_vecs:
        # If no words in the sentence are in our vocab, return a zero vector
        return np.zeros(word_vectors.vector_size)

    # Average the word vectors to get a single sentence vector
    sentence_vec = np.mean(word_vecs, axis=0)
    return sentence_vec

# Convert our text data into numerical vectors and labels
X_train_list = [sentence_to_vector(text, wv) for text, label in train_data]
y_train_list = [label for text, label in train_data]
print("\n--- X Train List ---")
print(train_data[0], "->", X_train_list[0])
print(train_data[1], "->", X_train_list[1])

# Convert lists to PyTorch Tensors
X_train = torch.tensor(np.array(X_train_list), dtype=torch.float32)
y_train = torch.tensor(np.array(y_train_list), dtype=torch.float32).view(-1, 1)

# --- Part 3: Define the PyTorch Model ---

class SentimentClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(SentimentClassifier, self).__init__()
        self.layer1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_dim, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.layer1(x)
        out = self.relu(out)
        out = self.layer2(out)
        out = self.sigmoid(out)
        return out

# Model parameters
INPUT_DIM = wv.vector_size  # Should be 300
HIDDEN_DIM = 64
OUTPUT_DIM = 1

model = SentimentClassifier(INPUT_DIM, HIDDEN_DIM, OUTPUT_DIM)

# --- Part 4: Train the Model ---

criterion = nn.BCELoss()  # Binary Cross-Entropy Loss for binary classification
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 100
for epoch in range(num_epochs):
    model.train()

    # Forward pass
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# --- Part 5: Test the Model ---

print("\n--- TESTING THE MODEL ---")
model.eval()
with torch.no_grad():
    # Test with a new positive sentence
    test_positive_text = "This was an amazing and wonderful film."
    test_vec = torch.tensor(sentence_to_vector(test_positive_text, wv), dtype=torch.float32)
    prediction = model(test_vec)
    print(f"Sentence: '{test_positive_text}'")
    print(f"Prediction: {prediction.item():.4f} -> {'Positive' if prediction.item() > 0.5 else 'Negative'}")

    # Test with a new negative sentence
    test_negative_text = "The movie was awful and I hated it."
    test_vec = torch.tensor(sentence_to_vector(test_negative_text, wv), dtype=torch.float32)
    prediction = model(test_vec)
    print(f"Sentence: '{test_negative_text}'")
    print(f"Prediction: {prediction.item():.4f} -> {'Positive' if prediction.item() > 0.5 else 'Negative'}")
```

## Day 7 | Sequences with RNNs and LSTMs

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import Counter

# --- Part 1: Prepare Data and Vocabulary ---

# Our simple dataset [("sentence", label), ...]
# 0 for Negative, 1 for Positive
train_data = [
    ("this movie was terrible", 0),
    ("i absolutely loved this", 1),
    ("the acting was poor", 0),
    ("this is the best film i have seen", 1),
    ("a truly brilliant masterpiece", 1),
    ("what a horrible experience", 0)
]

# Create vocabulary
word_counts = Counter()
for text, _ in train_data:
    word_counts.update(text.split())

# Assign an integer index to each word.
# <PAD> is a special token for padding. <UNK> for unknown words.
vocab = {"<PAD>": 0, "<UNK>": 1}
for word, _ in word_counts.items():
    vocab[word] = len(vocab)
vocab_size = len(vocab)

def text_to_sequence(text, vocab):
    """Converts a sentence into a sequence of integer IDs."""
    return [vocab.get(word, vocab["<UNK>"]) for word in text.split()]

# --- Part 2: Define the PyTorch LSTM Model ---

class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(LSTMClassifier, self).__init__()
        # This layer turns word IDs into dense vectors.
        # It can be initialized with pre-trained vectors or learn them from scratch.
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)

        # The LSTM layer which processes the sequence of embeddings.
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)

        # A standard fully-connected layer to map the LSTM output to our final prediction.
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # x shape: (batch_size, seq_length)
        embedded = self.embedding(x)
        # embedded shape: (batch_size, seq_length, embedding_dim)

        # We only need the final hidden state of the LSTM
        # lstm_out shape: (batch_size, seq_length, hidden_dim)
        # self.hidden shape: (1, batch_size, hidden_dim)
        lstm_out, (hidden, cell) = self.lstm(embedded)

        # We take the hidden state from the last time step
        final_hidden_state = hidden.squeeze(0)

        out = self.fc(final_hidden_state)
        out = self.sigmoid(out)
        return out

# Model parameters
EMBEDDING_DIM = 50
HIDDEN_DIM = 32
OUTPUT_DIM = 1
model = LSTMClassifier(vocab_size, EMBEDDING_DIM, HIDDEN_DIM, OUTPUT_DIM)

# --- Part 3: Train the Model ---

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Convert data for training
sequences = [text_to_sequence(text, vocab) for text, label in train_data]
labels = [label for _, label in train_data]

# Pad sequences to the same length for batching
# In a real scenario, we'd do this in batches, but for simplicity, we do it all at once.
max_len = max(len(seq) for seq in sequences)
padded_sequences = torch.zeros((len(sequences), max_len), dtype=torch.long)
for i, seq in enumerate(sequences):
    padded_sequences[i, :len(seq)] = torch.tensor(seq)

X_train = padded_sequences
y_train = torch.tensor(labels, dtype=torch.float32).view(-1, 1)

# Training loop
num_epochs = 200
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# --- Part 4: Test the Model ---
print("\n--- TESTING THE MODEL ---")
model.eval()
with torch.no_grad():
    test_text = "this was a fantastic and amazing film"
    test_seq = torch.tensor([text_to_sequence(test_text, vocab)], dtype=torch.long)
    prediction = model(test_seq)
    print(f"Sentence: '{test_text}'")
    print(f"Prediction: {prediction.item():.4f} -> {'Positive' if prediction.item() > 0.5 else 'Negative'}")
```

## Day 8 | Gated Recurrent Units (GRUs) and Bidirectionality

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import Counter

# --- Part 1: Data and Vocab (Identical to Day 7) ---
train_data = [
    ("this movie was terrible", 0),
    ("i absolutely loved this", 1),
    ("the acting was poor", 0),
    ("this is the best film i have seen", 1),
    ("a truly brilliant masterpiece", 1),
    ("what a horrible experience", 0)
]
word_counts = Counter()
for text, _ in train_data:
    word_counts.update(text.split())
vocab = {"<PAD>": 0, "<UNK>": 1}
for word, _ in word_counts.items():
    vocab[word] = len(vocab)
vocab_size = len(vocab)

def text_to_sequence(text, vocab):
    return [vocab.get(word, vocab["<UNK>"]) for word in text.split()]

# --- Part 2: Define the Bidirectional GRU Model ---

class BiGRUClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(BiGRUClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)

        # The key changes are here:
        # 1. We use nn.GRU instead of nn.LSTM.
        # 2. We set `bidirectional=True`.
        self.gru = nn.GRU(embedding_dim,
                          hidden_dim,
                          batch_first=True,
                          bidirectional=True) # <-- The magic happens here

        # The input to the fully-connected layer is now twice the hidden_dim
        # because we concatenate the forward and backward hidden states.
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        embedded = self.embedding(x)
        gru_out, hidden = self.gru(embedded)

        # `hidden` now has a shape of (num_directions * num_layers, batch_size, hidden_dim)
        # For a single-layer bidirectional GRU, this is (2, batch_size, hidden_dim).
        # We concatenate the final forward (hidden[-2,:,:]) and backward (hidden[-1,:,:]) hidden states.
        hidden_cat = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)

        out = self.fc(hidden_cat)
        out = self.sigmoid(out)
        return out

# --- Part 3 & 4: Training and Testing (Identical to Day 7) ---

# Model parameters
EMBEDDING_DIM = 50
HIDDEN_DIM = 32 # This is the size for one direction
OUTPUT_DIM = 1
model = BiGRUClassifier(vocab_size, EMBEDDING_DIM, HIDDEN_DIM, OUTPUT_DIM)

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Prepare data
sequences = [text_to_sequence(text, vocab) for text, label in train_data]
labels = [label for _, label in train_data]
max_len = max(len(seq) for seq in sequences)
padded_sequences = torch.zeros((len(sequences), max_len), dtype=torch.long)
for i, seq in enumerate(sequences):
    padded_sequences[i, :len(seq)] = torch.tensor(seq)
X_train = padded_sequences
y_train = torch.tensor(labels, dtype=torch.float32).view(-1, 1)

# Training loop
num_epochs = 200
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

## Day 10 | Self-Attention Head

```python
import torch
import torch.nn as nn

class SelfAttention(nn.Module):
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads

        assert (
            self.head_dim * heads == embed_size
        ), "Embedding size needs to be divisible by heads"

        # Linear layers to project input into Query, Key, and Value spaces
        self.values = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.keys = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.queries = nn.Linear(self.head_dim, self.head_dim, bias=False)

        # A final linear layer to combine the outputs
        self.fc_out = nn.Linear(heads * self.head_dim, embed_size, bias=False)

    def forward(self, values, keys, query, mask):
        # Get number of training examples and sequence lengths
        N = query.shape[0]
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]

        # Reshape the input for multi-head attention
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        # values shape: (N, value_len, heads, head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        # keys shape: (N, key_len, heads, head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)
        # queries shape: (N, query_len, heads, head_dim)

        # Apply the linear projections
        values = self.values(values)
        keys = self.keys(keys)
        queries = self.queries(queries)

        # Permute Q, K, and V for matrix multiplication
        # We want to multiply (q_len, head_dim) with (key_len, head_dim)^T
        # so we rearrange dimensions to (N, heads, seq_len, head_dim)
        queries = queries.permute(0, 2, 1, 3)
        # queries shape: (N, heads, query_len, head_dim)
        keys = keys.permute(0, 2, 1, 3)
        # keys shape: (N, heads, key_len, head_dim)
        values = values.permute(0, 2, 1, 3)
        # values shape: (N, heads, value_len, head_dim)


        # --- ATTENTION CALCULATION ---

        # 1. Mat-mul Queries and Keys (transposed)
        # Multiply queries (N, H, q_len, D_h) with keys (N, H, k_len, D_h)
        # We transpose the last two dimensions of keys
        energy = torch.matmul(queries, keys.transpose(-2, -1))
        # energy shape: (N, heads, query_len, key_len)

        if mask is not None:
            energy = energy.masked_fill(mask == 0, float("-1e20"))

        # 2. Scale and apply Softmax
        attention = torch.softmax(energy / (self.embed_size ** (1 / 2)), dim=-1)
        # attention shape: (N, heads, query_len, key_len)

        # 3. Mat-mul attention scores with Values
        out = torch.matmul(attention, values)
        # out shape: (N, heads, query_len, head_dim)

        # Reshape back to the original embedding format
        # First, make dimensions contiguous by swapping heads and query_len
        out = out.permute(0, 2, 1, 3).contiguous()
        # out shape: (N, query_len, heads, head_dim)

        # Then, combine the last two dimensions (heads and head_dim)
        out = out.reshape(N, query_len, self.heads * self.head_dim)
        # out shape: (N, query_len, embed_size)

        # Pass through the final fully connected layer
        out = self.fc_out(out)
        # out shape: (N, query_len, embed_size)

        return out

# --- Let's test it ---
embed_size = 256
heads = 1
seq_length = 10
batch_size = 3

x = torch.rand(batch_size, seq_length, embed_size)

self_attention = SelfAttention(embed_size, heads)
output = self_attention(x, x, x, mask=None)

print("--- SELF-ATTENTION (no einsum) ---")
print("Input shape:", x.shape)
print("Output shape:", output.shape)
print("Success! The output has the same shape as the input.")
```

## Day 11 | Multi-Head Attention

```python
import torch
import torch.nn as nn

class SelfAttention(nn.Module):
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads

        assert (
            self.head_dim * heads == embed_size
        ), "Embedding size must be divisible by heads"

        # These linear layers will be applied to each head
        self.values = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.keys = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.queries = nn.Linear(self.head_dim, self.head_dim, bias=False)

        # This layer combines the concatenated outputs of all heads
        self.fc_out = nn.Linear(heads * self.head_dim, embed_size, bias=False)

    def forward(self, values, keys, query, mask):
        N = query.shape[0]
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]

        # Split the embedding into self.heads different pieces
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)

        # Apply linear projections to each head
        values = self.values(values)
        keys = self.keys(keys)
        queries = self.queries(queries)

        # Permute for matrix multiplication
        queries = queries.permute(0, 2, 1, 3)
        keys = keys.permute(0, 2, 1, 3)
        values = values.permute(0, 2, 1, 3)

        # Scaled dot-product attention for all heads in parallel
        energy = torch.matmul(queries, keys.transpose(-2, -1))
        attention = torch.softmax(energy / (self.embed_size ** (1 / 2)), dim=-1)
        out = torch.matmul(attention, values)

        # Concatenate heads
        out = out.permute(0, 2, 1, 3).contiguous()
        out = out.reshape(N, query_len, self.heads * self.head_dim)

        # Pass through the final linear layer
        out = self.fc_out(out)
        return out


# --- Let's test it with MULTIPLE heads ---
embed_size = 256
heads = 8 # The only change from yesterday!
seq_length = 10
batch_size = 3

# The head_dim is now 256 / 8 = 32.
# This means we will have 8 parallel attention calculations
# each operating on smaller vectors of size 32.

x = torch.rand(batch_size, seq_length, embed_size)

# Create an instance of our self-attention module with 8 heads
multi_head_attention = SelfAttention(embed_size, heads)

output = multi_head_attention(x, x, x, mask=None)

print("--- MULTI-HEAD ATTENTION ---")
print("Input shape:", x.shape)
print("Output shape:", output.shape)
print("Success! The output shape is the same, but it now contains combined information from 8 different attention heads.")
```

## Day 12 | Transformer Block

```python
import torch
import torch.nn as nn

# Assume the SelfAttention class from Day 11 is defined here or imported
class SelfAttention(nn.Module):
    # --- Paste the full SelfAttention class code from Day 11 here ---
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads
        assert (self.head_dim * heads == embed_size), "Embedding size must be divisible by heads"
        self.values = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.keys = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.queries = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.fc_out = nn.Linear(heads * self.head_dim, embed_size, bias=False)
    def forward(self, values, keys, query, mask):
        N = query.shape[0]
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)
        values = self.values(values)
        keys = self.keys(keys)
        queries = self.queries(queries)
        queries = queries.permute(0, 2, 1, 3)
        keys = keys.permute(0, 2, 1, 3)
        values = values.permute(0, 2, 1, 3)
        energy = torch.matmul(queries, keys.transpose(-2, -1))
        attention = torch.softmax(energy / (self.embed_size ** (1 / 2)), dim=-1)
        out = torch.matmul(attention, values)
        out = out.permute(0, 2, 1, 3).contiguous()
        out = out.reshape(N, query_len, self.heads * self.head_dim)
        out = self.fc_out(out)
        return out
```

```python
class TransformerBlock(nn.Module):
    def __init__(self, embed_size, heads, dropout, forward_expansion):
        super(TransformerBlock, self).__init__()
        self.attention = SelfAttention(embed_size, heads)

        # Two Layer Normalization layers
        self.norm1 = nn.LayerNorm(embed_size)
        self.norm2 = nn.LayerNorm(embed_size)

        # The Feed-Forward Network
        self.feed_forward = nn.Sequential(
            nn.Linear(embed_size, forward_expansion * embed_size),
            nn.ReLU(),
            nn.Linear(forward_expansion * embed_size, embed_size)
        )

        self.dropout = nn.Dropout(dropout)

    def forward(self, value, key, query, mask):
        # 1. Attention sublayer
        attention_out = self.attention(value, key, query, mask)

        # 2. Add & Norm
        # Add the original query (residual connection) and then apply dropout and normalization
        x = self.dropout(attention_out) + query
        x = self.norm1(x)

        # 3. Feed-Forward sublayer
        forward_out = self.feed_forward(x)

        # 4. Add & Norm
        # Add the result from the previous Add & Norm layer and then apply dropout and normalization
        out = self.dropout(forward_out) + x
        out = self.norm2(out)

        return out

# --- Let's test it ---
embed_size = 256
heads = 8
dropout = 0.1
forward_expansion = 4 # In the paper, the FFN hidden layer is 4x the embed_size

# Create an instance of our Transformer block
transformer_block = TransformerBlock(embed_size, heads, dropout, forward_expansion)

# Create a sample input
batch_size = 3
seq_length = 10
x = torch.rand(batch_size, seq_length, embed_size)

# Pass the input through the block
output = transformer_block(x, x, x, mask=None)

print("--- TRANSFORMER BLOCK ---")
print("Input shape:", x.shape)
print("Output shape:", output.shape)
print("Success! The Transformer block processes the input without changing its shape.")
```

## Day 13 | Transformer Encoder

```python
import torch
import torch.nn as nn
import math

# Assume the SelfAttention and TransformerBlock classes are defined here
class SelfAttention(nn.Module):
    # --- Paste the full SelfAttention class code from Day 11 here ---
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        self.heads = heads
        self.head_dim = embed_size // heads
        assert (self.head_dim * heads == embed_size), "Embedding size must be divisible by heads"
        self.values = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.keys = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.queries = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.fc_out = nn.Linear(heads * self.head_dim, embed_size, bias=False)
    def forward(self, values, keys, query, mask):
        N = query.shape[0]
        value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)
        values = self.values(values)
        keys = self.keys(keys)
        queries = self.queries(queries)
        queries = queries.permute(0, 2, 1, 3)
        keys = keys.permute(0, 2, 1, 3)
        values = values.permute(0, 2, 1, 3)
        energy = torch.matmul(queries, keys.transpose(-2, -1))
        attention = torch.softmax(energy / (self.embed_size ** (1 / 2)), dim=-1)
        out = torch.matmul(attention, values)
        out = out.permute(0, 2, 1, 3).contiguous()
        out = out.reshape(N, query_len, self.heads * self.head_dim)
        out = self.fc_out(out)
        return out

```

```python

class TransformerBlock(nn.Module):
    # --- Paste the full TransformerBlock class code from Day 12 here ---
    def __init__(self, embed_size, heads, dropout, forward_expansion):
        super(TransformerBlock, self).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm1 = nn.LayerNorm(embed_size)
        self.norm2 = nn.LayerNorm(embed_size)
        self.feed_forward = nn.Sequential(
            nn.Linear(embed_size, forward_expansion * embed_size),
            nn.ReLU(),
            nn.Linear(forward_expansion * embed_size, embed_size)
        )
        self.dropout = nn.Dropout(dropout)
    def forward(self, value, key, query, mask):
        attention_out = self.attention(value, key, query, mask)
        x = self.dropout(attention_out) + query
        x = self.norm1(x)
        forward_out = self.feed_forward(x)
        out = self.dropout(forward_out) + x
        out = self.norm2(out)
        return out
```

```python
# The final Encoder module
class Encoder(nn.Module):
    def __init__(
        self,
        src_vocab_size,
        embed_size,
        num_layers,
        heads,
        device,
        forward_expansion,
        dropout,
        max_length,
    ):
        super(Encoder, self).__init__()
        self.embed_size = embed_size
        self.device = device
        self.word_embedding = nn.Embedding(src_vocab_size, embed_size)
        # Create a positional encoding matrix
        pe = torch.zeros(max_length, embed_size)
        position = torch.arange(0, max_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_size, 2).float() * (-math.log(10000.0) / embed_size))

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0) # Add a batch dimension

        # Register 'pe' as a buffer. It's part of the model's state,
        # but not a parameter that should be trained.
        self.register_buffer('pe', pe)

        self.layers = nn.ModuleList(
            [
                TransformerBlock(
                    embed_size,
                    heads,
                    dropout=dropout,
                    forward_expansion=forward_expansion,
                )
                for _ in range(num_layers)
            ]
        )

        self.dropout = nn.Dropout(dropout)

    def forward(self, x, mask):
        N, seq_length = x.shape
        # Get the word embeddings
        word_embed = self.word_embedding(x)

        # Add the positional encoding slice to the word embeddings
        # self.pe is (1, max_len, embed_size). We slice it to match the input seq_length.
        pos_embed = self.pe[:, :seq_length].to(self.device)
        out = self.dropout(word_embed + pos_embed)

        # Pass through each Transformer block
        for layer in self.layers:
            out = layer(out, out, out, mask)

        return out

# --- Let's test it ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create a sample input of token IDs (not raw text)
# This would be the output of our tokenizer
x = torch.randint(0, 1000, (3, 10)).to(device) # (batch_size, seq_length)

# Encoder parameters
src_vocab_size = 1000
embed_size = 256
num_layers = 6 # The paper used 6 layers
heads = 8
forward_expansion = 4
dropout = 0.1
max_length = 100

# Create an instance of our full Encoder
encoder = Encoder(
    src_vocab_size,
    embed_size,
    num_layers,
    heads,
    device,
    forward_expansion,
    dropout,
    max_length,
).to(device)

# Pass the input through the encoder
# We'll learn about masking later
output = encoder(x, mask=None)

print("--- FULL TRANSFORMER ENCODER ---")
print("Input shape (token IDs):", x.shape)
print("Output shape (contextualized embeddings):", output.shape)
print(f"Success! The encoder output has shape (batch_size, seq_length, embed_size).")
```

## Day 14 | Mask

```python
import torch

def create_look_ahead_mask(size):
    """
    Creates a square matrix with 1s on the lower triangle and 0s on the upper triangle.
    The upper triangle (where row < col) represents future positions we want to mask.

    Args:
        size (int): The length of the sequence.

    Returns:
        torch.Tensor: A mask of shape (1, 1, size, size).
    """
    # Create a square matrix of ones. Shape: (size, size)
    mask = torch.ones(size, size)

    # Create an upper triangular matrix from our square matrix.
    # The `triu` function sets all elements on the main diagonal and below it to 0.
    # We use `diagonal=1` to keep the main diagonal (a token can attend to itself).
    mask = torch.triu(mask, diagonal=1)

    # The `masked_fill` in PyTorch expects 0s where we want to mask and 1s otherwise.
    # So we invert our logic. Or more simply, create a lower triangular matrix.

    # A simpler way:
    mask = torch.tril(torch.ones(size, size)) # Lower triangular matrix of ones

    # The SelfAttention module will expect a boolean mask where `False` means mask.
    # Or a float mask where `-inf` means mask.
    # Let's create the boolean version.

    # Final correct implementation:
    # A mask where True means keep and False means hide.
    mask = torch.tril(torch.ones(size, size)) == 1
    return mask


# --- Let's test it ---
seq_length = 5
mask = create_look_ahead_mask(seq_length)

print(f"--- LOOK-AHEAD MASK (size={seq_length}) ---")
print(mask)
print("\nExplanation:")
print("At row 0 (predicting 1st word), we can only see column 0 (the 1st word).")
print("At row 2 (predicting 3rd word), we can see columns 0, 1, and 2 (the 1st, 2nd, and 3rd words).")
print("It can never see future words (the upper triangle is all False).")
```

## Day 15 | Decoder Block

```python

class DecoderBlock(nn.Module):
    def __init__(self, embed_size, heads, forward_expansion, dropout, device):
        super(DecoderBlock, self).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm = nn.LayerNorm(embed_size)

        # The new Encoder-Decoder attention layer
        self.transformer_block = TransformerBlock(
            embed_size, heads, dropout, forward_expansion
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, value, key, src_mask, trg_mask):
        # 1. Masked Multi-Head Self-Attention
        # This is where the decoder looks at the sequence it has generated so far.
        # The `trg_mask` (look-ahead mask) is applied here.
        attention = self.attention(x, x, x, trg_mask)
        query = self.dropout(self.norm(attention + x))

        # 2. Encoder-Decoder Attention
        # This is where the decoder looks at the encoded input sentence.
        # Q comes from the decoder, K and V come from the encoder.
        # `src_mask` is used here to mask padding tokens from the source sentence.
        out = self.transformer_block(value, key, query, src_mask)
        return out

# NOTE: The above DecoderBlock is a simplified implementation.
# A more faithful one would separate the Encoder-Decoder attention and FFN.
# Let's write that for clarity.

class DecoderBlockFaithful(nn.Module):
    def __init__(self, embed_size, heads, forward_expansion, dropout, device):
        super(DecoderBlockFaithful, self).__init__()
        self.norm = nn.LayerNorm(embed_size)
        self.attention = SelfAttention(embed_size, heads) # Masked self-attention
        self.encoder_attention = SelfAttention(embed_size, heads) # Encoder-decoder attention

        self.feed_forward = nn.Sequential(
            nn.Linear(embed_size, forward_expansion * embed_size),
            nn.ReLU(),
            nn.Linear(forward_expansion * embed_size, embed_size)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, value, key, src_mask, trg_mask):
        # 1. Masked Self-Attention
        attention = self.attention(x, x, x, trg_mask)
        # Add & Norm
        query = self.dropout(self.norm(attention + x))

        # 2. Encoder-Decoder Attention
        # Q=query (from decoder), K=key (from encoder), V=value (from encoder)
        out = self.encoder_attention(value, key, query, src_mask)
        # Add & Norm
        x = self.dropout(self.norm(out + query))

        # 3. Feed-Forward
        forward = self.feed_forward(x)
        # Add & Norm
        out = self.dropout(self.norm(forward + x))
        return out


# --- Let's test the faithful implementation ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Decoder block parameters
embed_size = 256
heads = 8
forward_expansion = 4
dropout = 0.1

# Dummy inputs
batch_size = 3
target_seq_len = 12
source_seq_len = 10

# The decoder's input so far (e.g., embeddings of the translated sentence)
x = torch.rand(batch_size, target_seq_len, embed_size).to(device)
# The output from the final layer of the Encoder
enc_out = torch.rand(batch_size, source_seq_len, embed_size).to(device)

# The look-ahead mask for the target sequence
trg_mask = torch.tril(torch.ones((target_seq_len, target_seq_len))).to(device)

# Create an instance of our decoder block
decoder_block = DecoderBlockFaithful(embed_size, heads, forward_expansion, dropout, device).to(device)

# Pass the inputs through
output = decoder_block(x, enc_out, enc_out, src_mask=None, trg_mask=trg_mask)

print("--- TRANSFORMER DECODER BLOCK ---")
print("Decoder input shape (x):", x.shape)
print("Encoder output shape (enc_out):", enc_out.shape)
print("Final output shape:", output.shape)
print("Success! The output shape is (batch_size, target_seq_len, embed_size).")
```

## Day 16 | Decoder

```python

# The final Decoder module
class Decoder(nn.Module):
    def __init__(
        self,
        trg_vocab_size,
        embed_size,
        num_layers,
        heads,
        forward_expansion,
        dropout,
        device,
        max_length,
    ):
        super(Decoder, self).__init__()
        self.device = device
        self.word_embedding = nn.Embedding(trg_vocab_size, embed_size)
        self.position_embedding = nn.Embedding(max_length, embed_size)

        self.layers = nn.ModuleList(
            [
                DecoderBlockFaithful(embed_size, heads, forward_expansion, dropout, device)
                for _ in range(num_layers)
            ]
        )

        # The crucial final layer
        self.fc_out = nn.Linear(embed_size, trg_vocab_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, enc_out, src_mask, trg_mask):
        N, seq_length = x.shape
        positions = torch.arange(0, seq_length).expand(N, seq_length).to(self.device)
        x = self.dropout(self.word_embedding(x) + self.position_embedding(positions))

        for layer in self.layers:
            x = layer(x, enc_out, enc_out, src_mask, trg_mask)

        # Pass through the final linear layer to get scores for each word in the vocab
        out = self.fc_out(x)
        return out

# --- Let's test it ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Decoder parameters
trg_vocab_size = 2000
embed_size = 256
num_layers = 6
heads = 8
forward_expansion = 4
dropout = 0.1
max_length = 100

# Dummy inputs
batch_size = 3
target_seq_len = 12
source_seq_len = 10

# The decoder's input (target token IDs)
x = torch.randint(0, trg_vocab_size, (batch_size, target_seq_len)).to(device)
# The output from the final layer of the Encoder
enc_out = torch.rand(batch_size, source_seq_len, embed_size).to(device)

# The look-ahead mask for the target sequence
trg_mask = torch.tril(torch.ones((target_seq_len, target_seq_len))).to(device)

# Create an instance of our full Decoder
decoder = Decoder(
    trg_vocab_size,
    embed_size,
    num_layers,
    heads,
    forward_expansion,
    dropout,
    device,
    max_length
).to(device)

# Pass the inputs through
output = decoder(x, enc_out, src_mask=None, trg_mask=trg_mask)

print("--- FULL TRANSFORMER DECODER ---")
print("Decoder input shape (token IDs):", x.shape)
print("Encoder output shape (enc_out):", enc_out.shape)
print("Final output shape (logits):", output.shape)
print(f"Success! The final output shape is (batch_size, target_seq_len, trg_vocab_size).")
```

## Day 17 | The Final Assembly - The Complete Transformer

```python
import torch
import torch.nn as nn

# All the component classes from Day 9 to Day 16 go here.
# For brevity, I am collapsing them. Make sure you have them in your script.

class SelfAttention(nn.Module):
    def __init__(self, embed_size, heads):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size; self.heads = heads; self.head_dim = embed_size // heads
        assert (self.head_dim * heads == embed_size), "Embed size needs to be divisible by heads"
        self.values = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.keys = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.queries = nn.Linear(self.head_dim, self.head_dim, bias=False)
        self.fc_out = nn.Linear(heads * self.head_dim, embed_size, bias=False)
    def forward(self, values, keys, query, mask):
        N = query.shape[0]; value_len, key_len, query_len = values.shape[1], keys.shape[1], query.shape[1]
        values = values.reshape(N, value_len, self.heads, self.head_dim)
        keys = keys.reshape(N, key_len, self.heads, self.head_dim)
        queries = query.reshape(N, query_len, self.heads, self.head_dim)
        values = self.values(values); keys = self.keys(keys); queries = self.queries(queries)
        queries = queries.permute(0, 2, 1, 3); keys = keys.permute(0, 2, 1, 3); values = values.permute(0, 2, 1, 3)
        energy = torch.matmul(queries, keys.transpose(-2, -1))
        if mask is not None: energy = energy.masked_fill(mask == 0, float("-1e20"))
        attention = torch.softmax(energy / (self.embed_size ** (1 / 2)), dim=-1)
        out = torch.matmul(attention, values)
        out = out.permute(0, 2, 1, 3).contiguous()
        out = out.reshape(N, query_len, self.heads * self.head_dim)
        out = self.fc_out(out)
        return out

class TransformerBlock(nn.Module):
    def __init__(self, embed_size, heads, dropout, forward_expansion):
        super(TransformerBlock, self).__init__()
        self.attention = SelfAttention(embed_size, heads)
        self.norm1 = nn.LayerNorm(embed_size); self.norm2 = nn.LayerNorm(embed_size)
        self.feed_forward = nn.Sequential(nn.Linear(embed_size, forward_expansion * embed_size), nn.ReLU(), nn.Linear(forward_expansion * embed_size, embed_size))
        self.dropout = nn.Dropout(dropout)
    def forward(self, value, key, query, mask):
        attention_out = self.attention(value, key, query, mask)
        x = self.dropout(attention_out) + query
        x = self.norm1(x)
        forward_out = self.feed_forward(x)
        out = self.dropout(forward_out) + x
        out = self.norm2(out)
        return out

class Encoder(nn.Module):
    def __init__(self, src_vocab_size, embed_size, num_layers, heads, device, forward_expansion, dropout, max_length):
        super(Encoder, self).__init__()
        self.embed_size = embed_size; self.device = device
        self.word_embedding = nn.Embedding(src_vocab_size, embed_size)
        self.position_embedding = nn.Embedding(max_length, embed_size)
        self.layers = nn.ModuleList([TransformerBlock(embed_size, heads, dropout=dropout, forward_expansion=forward_expansion) for _ in range(num_layers)])
        self.dropout = nn.Dropout(dropout)
    def forward(self, x, mask):
        N, seq_length = x.shape
        positions = torch.arange(0, seq_length).expand(N, seq_length).to(self.device)
        out = self.dropout(self.word_embedding(x) + self.position_embedding(positions))
        for layer in self.layers: out = layer(out, out, out, mask)
        return out

class DecoderBlock(nn.Module):
    def __init__(self, embed_size, heads, forward_expansion, dropout, device):
        super(DecoderBlock, self).__init__()
        self.norm = nn.LayerNorm(embed_size)
        self.attention = SelfAttention(embed_size, heads)
        self.encoder_attention = SelfAttention(embed_size, heads)
        self.feed_forward = nn.Sequential(nn.Linear(embed_size, forward_expansion * embed_size), nn.ReLU(), nn.Linear(forward_expansion * embed_size, embed_size))
        self.dropout = nn.Dropout(dropout)
    def forward(self, x, value, key, src_mask, trg_mask):
        attention = self.attention(x, x, x, trg_mask)
        query = self.dropout(self.norm(attention + x))
        out = self.encoder_attention(value, key, query, src_mask)
        x = self.dropout(self.norm(out + query))
        forward = self.feed_forward(x)
        out = self.dropout(self.norm(forward + x))
        return out

class Decoder(nn.Module):
    def __init__(self, trg_vocab_size, embed_size, num_layers, heads, forward_expansion, dropout, device, max_length):
        super(Decoder, self).__init__()
        self.device = device
        self.word_embedding = nn.Embedding(trg_vocab_size, embed_size)
        self.position_embedding = nn.Embedding(max_length, embed_size)
        self.layers = nn.ModuleList([DecoderBlock(embed_size, heads, forward_expansion, dropout, device) for _ in range(num_layers)])
        self.fc_out = nn.Linear(embed_size, trg_vocab_size)
        self.dropout = nn.Dropout(dropout)
    def forward(self, x, enc_out, src_mask, trg_mask):
        N, seq_length = x.shape
        positions = torch.arange(0, seq_length).expand(N, seq_length).to(self.device)
        x = self.dropout(self.word_embedding(x) + self.position_embedding(positions))
        for layer in self.layers: x = layer(x, enc_out, enc_out, src_mask, trg_mask)
        out = self.fc_out(x)
        return out


# --- The Final Transformer Class ---
class Transformer(nn.Module):
    def __init__(
        self,
        src_vocab_size,
        trg_vocab_size,
        src_pad_idx,
        trg_pad_idx,
        embed_size=256,
        num_layers=6,
        forward_expansion=4,
        heads=8,
        dropout=0.1,
        device="cpu",
        max_length=100
    ):
        super(Transformer, self).__init__()

        self.encoder = Encoder(
            src_vocab_size, embed_size, num_layers, heads, device,
            forward_expansion, dropout, max_length
        )

        self.decoder = Decoder(
            trg_vocab_size, embed_size, num_layers, heads, forward_expansion,
            dropout, device, max_length
        )

        self.src_pad_idx = src_pad_idx
        self.trg_pad_idx = trg_pad_idx
        self.device = device

    def make_src_mask(self, src):
        # src_mask shape: (N, 1, 1, src_len)
        src_mask = (src != self.src_pad_idx).unsqueeze(1).unsqueeze(2)
        return src_mask.to(self.device)

    def make_trg_mask(self, trg):
        N, trg_len = trg.shape
        # trg_mask shape: (N, 1, trg_len, trg_len)
        trg_mask = torch.tril(torch.ones((trg_len, trg_len))).expand(
            N, 1, trg_len, trg_len
        )
        return trg_mask.to(self.device)

    def forward(self, src, trg):
        src_mask = self.make_src_mask(src)
        trg_mask = self.make_trg_mask(trg)

        enc_out = self.encoder(src, src_mask)
        out = self.decoder(trg, enc_out, src_mask, trg_mask)
        return out

# --- Let's test it ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dummy inputs
x = torch.randint(1, 1000, (3, 10)).to(device) # source sentence
trg = torch.randint(1, 2000, (3, 12)).to(device) # target sentence

src_pad_idx = 0
trg_pad_idx = 0
src_vocab_size = 1000
trg_vocab_size = 2000

model = Transformer(src_vocab_size, trg_vocab_size, src_pad_idx, trg_pad_idx, device=device).to(device)

# Pass the inputs through
output = model(x, trg[:, :-1]) # We don't feed the last token of target to decoder

print("--- COMPLETE TRANSFORMER MODEL ---")
print("Input shape (src):", x.shape)
print("Input shape (trg):", trg[:, :-1].shape)
print("Final output shape (logits):", output.shape)
print("Success! You have built a complete Transformer from scratch.")
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import numpy as np
import evaluate

# 1. Load the dataset and tokenizer (same as yesterday)
print("Loading data...")
imdb_dataset = load_dataset("imdb")
model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_datasets = imdb_dataset.map(tokenize_function, batched=True)

# Use smaller subsets for a quick training example
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
print("Data loaded and prepared.")

# 2. Load the pre-trained model
# `num_labels` tells the model we are doing binary classification.
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, num_labels=2)
print("Model loaded.")

```

```python
# 3. Define Training Arguments
# This class contains all the hyperparameters for training.
training_args = TrainingArguments(
    output_dir="my_awesome_model",      # Directory to save the model
    eval_strategy="epoch",       # Evaluate every epoch
    num_train_epochs=3,                # Number of training epochs
    per_device_train_batch_size=8,     # Batch size for training
    per_device_eval_batch_size=8,      # Batch size for evaluation
    warmup_steps=500,                  # Number of warmup steps for learning rate scheduler
    weight_decay=0.01,                 # Strength of weight decay
    logging_dir='./logs',              # Directory for storing logs
    logging_steps=10,
    report_to="none",
)

```

```python
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
from transformers import pipeline

# --- Part 1: The Retriever (Same as Day 22) ---

documents = [
    "The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.",
    "The Great Wall of China is a series of fortifications made of stone, brick, and other materials.",
    "Lionel Messi is a professional footballer who plays as a forward for Inter Miami.",
    "The first programmable computer, the Z1, was created by Konrad Zuse in Germany between 1936 and 1938.",
    "The capital of Armenia is Yerevan, one of the world's oldest continuously inhabited cities."
]

# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')
document_embeddings = model.encode(documents)

# Build FAISS index
d = document_embeddings.shape[1]
index = faiss.IndexFlatL2(d)
index.add(document_embeddings)

# --- Part 2: The Generator ---

# Load a text-generation pipeline
print("Loading text-generation pipeline...")
generator = pipeline('text2text-generation', model='google/flan-t5-base')
print("Pipeline loaded.")


# --- Part 3: The Combined RAG Function ---

def answer_question(query):
    # 1. Retrieve
    query_embedding = model.encode([query])
    k = 1 # Retrieve the single most relevant document
    distances, indices = index.search(query_embedding, k)

    retrieved_document = documents[indices[0][0]]

    # 2. Augment
    # Create the prompt with the retrieved context
    prompt = f"""
    Based on the following context, answer the question.

    Context:
    {retrieved_document}

    Question:
    {query}

    Answer:
    """

    # 3. Generate
    result = generator(prompt, max_length=50)

    print(f"Query: {query}")
    print(f"Retrieved Document: '{retrieved_document}'")
    print(f"Generated Answer: {result[0]['generated_text']}")
    print("-" * 20)

# --- Let's test our RAG system ---
answer_question("Where is the Eiffel Tower located?")
answer_question("Who created the first computer?")
answer_question("Who is the president of the United States?")
```

```python
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer, CrossEncoder
from transformers import pipeline

# --- Part 1: Setup the Retriever (Bi-Encoder + FAISS) ---

documents = [
    "The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France. It is named after the engineer Gustave Eiffel.",
    "The Great Wall of China is a series of fortifications made of stone, brick, tampered earth, wood, and other materials, generally built along an east-to-west line across the historical northern borders of China.",
    "Lionel Messi is a professional footballer who plays as a forward for Inter Miami. He is widely regarded as one of the greatest players of all time.",
    "The first programmable computer, the Z1, was created by Konrad Zuse in Germany. He is often regarded as the inventor of the computer.",
    "The capital of Armenia is Yerevan, one of the world's oldest continuously inhabited cities. It is situated along the Hrazdan River.",
    "Gustave Eiffel was a French civil engineer. His company designed and built the Eiffel Tower for the 1889 Universal Exposition."
]

# Load bi-encoder model for initial retrieval
retriever_model = SentenceTransformer('all-MiniLM-L6-v2')
document_embeddings = retriever_model.encode(documents)

# Build FAISS index
d = document_embeddings.shape[1]
index = faiss.IndexFlatL2(d)
index.add(document_embeddings)


# --- Part 2: Setup the Re-Ranker (Cross-Encoder) ---
reranker_model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')


# --- Part 3: Setup the Generator (LLM) ---
generator = pipeline('text2text-generation', model='google/flan-t5-base')


# --- Part 4: The Combined RAG Function with Re-Ranking ---

def answer_question_with_reranker(query, top_k_retrieval=4, top_k_rerank=2):
    print(f"Query: '{query}'")

    # 1. Retrieve
    # Get a larger pool of candidate documents from the fast retriever.
    query_embedding = retriever_model.encode([query])
    distances, indices = index.search(query_embedding, top_k_retrieval)

    retrieved_docs_text = [documents[i] for i in indices[0]]
    print(f"\nRetrieved {len(retrieved_docs_text)} candidate documents...")

    # 2. Re-Rank
    # Create pairs of [query, document] for the cross-encoder.
    sentence_pairs = [[query, doc] for doc in retrieved_docs_text]

    # Get new, more accurate scores.
    rerank_scores = reranker_model.predict(sentence_pairs)

    # Combine scores with documents and sort.
    scored_docs = list(zip(rerank_scores, retrieved_docs_text))
    scored_docs.sort(key=lambda x: x[0], reverse=True)

    print("\nRe-ranked scores:")
    for score, doc in scored_docs:
        print(f"  Score: {score:.4f} - Doc: '{doc[:80]}...'")

    # 3. Augment
    # Select the top documents after re-ranking to build the context.
    final_context = "\n".join([doc for score, doc in scored_docs[:top_k_rerank]])

    prompt = f"""
    Answer the following question based on the provided context.

    Context:
    {final_context}

    Question:
    {query}

    Answer:
    """

    # 4. Generate
    result = generator(prompt, max_length=100)
    print(f"\nFinal Generated Answer: {result[0]['generated_text']}")
    print("=" * 40 + "\n")


# --- Let's test our improved RAG system ---
answer_question_with_reranker("Who designed the Eiffel Tower?")
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain_huggingface import HuggingFacePipeline

# --- Part 1: Our Document Collection ---
# For this example, let's create a text file with our documents.
documents_text = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
The Great Wall of China is a series of fortifications made of stone, brick, and other materials.
Lionel Messi is a professional footballer who plays as a forward for Inter Miami.
The first programmable computer, the Z1, was created by Konrad Zuse in Germany between 1936 and 1938.
The capital of Armenia is Yerevan, one of the world's oldest continuously inhabited cities.
"""
with open("my_documents.txt", "w") as f:
    f.write(documents_text)


# --- Part 2: The LangChain RAG Pipeline ---

# 1. Load Documents
# Use a Document Loader to load data from a source.
loader = TextLoader('./my_documents.txt')
documents = loader.load()

# 2. Split Documents into Chunks
# Use a Text Splitter to create manageable chunks.
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0,separator="\n",)
docs = text_splitter.split_documents(documents)

# 3. Create Embeddings and Store in a Vector Store
# This single step creates embeddings and indexes them in FAISS.
embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
vectorstore = FAISS.from_documents(docs, embeddings)

# 4. Initialize the LLM
# We'll use the same model as before.
llm = HuggingFacePipeline.from_model_id(
    model_id="google/flan-t5-base",
    task="text2text-generation",
    pipeline_kwargs={"max_new_tokens": 100},
)

# 5. Create the RetrievalQA Chain
# This chain combines the retriever and the LLM.
# "stuff" is a simple method that crams all retrieved docs into the context.
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever() # Use our vector store as the retriever
)


# --- Part 3: Test the Chain ---
print("--- LangChain RAG Pipeline ---")
query = "Where is the Eiffel Tower located?"
result = qa_chain.invoke({"query": query})

print(f"Query: {query}")
print(f"Answer: {result['result']}")

print("-" * 20)

query = "Who created the first computer?"
result = qa_chain.invoke({"query": query})

print(f"Query: {query}")
print(f"Answer: {result['result']}")
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
from langchain_huggingface import HuggingFacePipeline

# --- Part 1: Setup the Vector Store (Same as Day 25) ---
documents_text = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
The Great Wall of China is a series of fortifications made of stone, brick, and other materials.
Lionel Messi is a professional footballer who plays as a forward for Inter Miami. He is widely regarded as one ofthe greatest players of all time.
The first programmable computer, the Z1, was created by Konrad Zuse in Germany between 1936 and 1938.
The capital of Armenia is Yerevan, one of the world's oldest continuously inhabited cities.
Gustave Eiffel was a French civil engineer. His company designed and built the Eiffel Tower for the 1889 Universal Exposition.
"""
with open("my_documents.txt", "w") as f:
    f.write(documents_text)

loader = TextLoader('./my_documents.txt')
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=150, chunk_overlap=20, separator="\n")
docs = text_splitter.split_documents(documents)
embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
vectorstore = FAISS.from_documents(docs, embeddings)

# --- Part 2: The Conversational Chain ---

# 1. Initialize the LLM
llm = HuggingFacePipeline.from_model_id(
    model_id="google/flan-t5-base",
    task="text2text-generation",
    pipeline_kwargs={"max_new_tokens": 100},
)

# 2. Setup Memory
# This memory object will store the conversation history.
# `return_messages=True` ensures it returns messages in a structured format.
memory = ConversationBufferMemory(
    memory_key='chat_history',
    return_messages=True
)

# 3. Create the ConversationalRetrievalChain
# This is the key component that combines the retriever, memory, and LLM.
conversation_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=vectorstore.as_retriever(),
    memory=memory
)


# --- Part 3: Test the Conversational Chatbot ---
print("--- LangChain Conversational RAG Chatbot ---")
print("Bot is ready. Type 'exit' to quit.")

while True:
    query = input("You: ")
    if query.lower() == 'exit':
        break

    result = conversation_chain.invoke({"question": query})
    print(f"Bot: {result['answer']}")
```

```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.tools import Tool
from langchain_experimental.tools.python.tool import PythonREPLTool
from langchain.agents import AgentExecutor, create_react_agent
from langchain import hub
from langchain_community.document_loaders import TextLoader

# Import necessary classes from transformers
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain_huggingface import HuggingFacePipeline


# --- Part 1: Setup the Retriever Tool (RAG System) ---
documents_text = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
The first programmable computer, the Z1, was created by Konrad Zuse in Germany between 1936 and 1938.
"""
with open("my_documents.txt", "w") as f:
    f.write(documents_text)

loader = TextLoader('./my_documents.txt')
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=150, chunk_overlap=0, separator='\n')
docs = text_splitter.split_documents(documents)
embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
vectorstore = FAISS.from_documents(docs, embeddings)

# --- THE FIX IS HERE: More robust LLM initialization ---
model_id = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id)

# Create a pipeline manually
pipe = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=200,
    # Explicitly set return_full_text to False to avoid issues with the streamer
    return_full_text=False,
)

# Wrap the pipeline in the LangChain object
llm = HuggingFacePipeline(pipeline=pipe)
# --- END OF FIX ---


# Create the RAG chain
rag_chain = RetrievalQA.from_chain_type(
    llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever()
)

# --- Part 2: Create the Tools ---
tools = [
    Tool(
        name="Document Knowledge Base",
        func=rag_chain.invoke,
        description="""Useful for when you need to answer questions about the Eiffel Tower or the first computer."""
    ),
    Tool(
        name="Calculator",
        func=PythonREPLTool().run,
        description="""Useful for when you need to answer questions about math or perform calculations."""
    ),
]


# --- Part 3: Create and Run the Agent ---
prompt = hub.pull("hwchase17/react")
agent = create_react_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# --- Test the agent ---
print("--- LangChain Agent with Tools ---")

# Test with a question for the knowledge base
agent_executor.invoke({"input": "Who created the first computer?"})

# Test with a question for the calculator
agent_executor.invoke({"input": "What is 25 raised to the power of 3?"})
```

```python
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)
from peft import get_peft_model, LoraConfig, TaskType
import numpy as np
import evaluate

# --- Part 1: Data Preparation ---
print("Loading and preparing data...")
imdb_dataset = load_dataset("imdb")
model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=512
    )

tokenized_datasets = imdb_dataset.map(tokenize_function, batched=True)
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
print("Data ready.")

# --- Part 2: Model Setup with PEFT ---
model = AutoModelForSequenceClassification.from_pretrained(
    model_checkpoint, num_labels=2
)
```

```python

peft_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,  # or SEQUENCE_CLASSIFICATION if on new peft
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.1,
    target_modules=["q_lin", "v_lin"],  # ✅ specify LoRA target layers
)
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()
print("Model ready for PEFT.")

# --- Part 3: Training ---
training_args = TrainingArguments(
    output_dir="my_lora_model",
    eval_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    logging_steps=50,
    report_to="none",  # Disable wandb
    save_strategy="epoch",
)

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)

print("Starting LoRA fine-tuning...")
trainer.train()
print("Fine-tuning complete.")

# Save only the LoRA adapter weights
model.save_pretrained("my_lora_adapter")
print("LoRA adapter saved to 'my_lora_adapter'")
```

```python
import torch
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer, BitsAndBytesConfig
from peft import get_peft_model, LoraConfig, prepare_model_for_kbit_training

# --- Part 1: Quantization Configuration ---
# This is where we configure the 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",      # Use a special 4-bit data type for weights
    bnb_4bit_compute_dtype=torch.bfloat16, # Use 16-bit floats for computation
    bnb_4bit_use_double_quant=True,  # Use a nested quantization for more efficiency
)

# --- Part 2: Load Model and Tokenizer ---
# We'll use a smaller, but still powerful, model for this example
# Note: For Llama models, you might need to authenticate with Hugging Face Hub first
model_name = "facebook/opt-1.3b" # A 1.3B parameter model, good for demonstration
print(f"Loading base model: {model_name}")

# Load the model with our quantization config
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto" # Automatically place the model on available devices (GPU)
)

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token # Set pad token for open-ended generation models


# --- Part 3: Prepare Model and Data for PEFT ---
# Prepare the model for k-bit training
model = prepare_model_for_kbit_training(model)

# Define the LoRA configuration
peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "v_proj"] # Target the query and value projections in attention
)

# Wrap the base model with get_peft_model
model = get_peft_model(model, peft_config)

model.print_trainable_parameters()
print("Model ready for QLoRA.")

# --- Part 4: Training (using a small text dataset) ---
# We'll use a small dataset for this demonstration
dataset = load_dataset("Abirate/english_quotes")
dataset = dataset["train"].train_test_split(test_size=0.1)["train"]

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["quote"], truncation=True, max_length=64, padding="max_length")

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["quote", "author", "tags"])
```

```python

training_args = TrainingArguments(
    output_dir="my_qlora_model",
    per_device_train_batch_size=4,
    num_train_epochs=1,
    logging_steps=20,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)

print("Starting QLoRA fine-tuning...")
trainer.train()
print("Fine-tuning complete.")
```
